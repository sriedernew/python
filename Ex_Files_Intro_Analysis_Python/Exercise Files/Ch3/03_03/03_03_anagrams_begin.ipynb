```python
wordclean = sorted(list(set([word.strip().lower() for word in open('words','r')])))
```

```python
wordclean[:10]
```

```python
sorted('magdeburg')
```

```python
sorted('magdeburg') == sorted('eburgmagd')
```

```python
def signature(word):
    return ''.join(sorted(word))
signature('MAGDEBURG')
```

```python
def anagram(myword):
    return [word for word in wordclean if signature(word) == signature(myword)]
anagram('bird')
```

```python
%timeit anagram('bird')
```

```python
words_bysing = {}
for word in wordclean:
    words_bysing[signature(word)].append(word)
```

```python
import collections
```

```python
words_bysing = collections.defaultdict(list)
for word in wordclean:
    words_bysing[signature(word)].append(word)
```

```python
words_bysing
```

```python
def anagram_fast(myword):
    return words_bysing[signature(myword)]
```

```python
anagram_fast('bird')
```

```python
anagram_fast('bird')
```

```python
anagram_fast('bird')
```

```python
anagrams_all = {word: anagram_fast(word) for word in wordclean if len(anagram_fast(word)) > 1}
```
